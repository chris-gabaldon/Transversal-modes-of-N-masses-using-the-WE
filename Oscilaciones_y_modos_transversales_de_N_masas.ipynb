# Ecuacion de Newton
En este punto se pide exhibir la  ecuacion de Newton que rige al sistema.Es importante notar que $\psi$ se refiere a la posicion de cada masa en el equilibrio. esta es una manera muy comoda de trabajar con numerosas masas y se puede demostrar que luego de cuentas es lo mismo que tomar un origen comun en la pared y trabajar con ese origen de coordenadas. No es el objetivo del TP asi que no lo haremos.

En el esquema del ejercicio tenemos una cadena de $N$  masas  que podrian ser distintas,separadas a una distancia $a$ , unidas por resortes de constante elastica $k$ y longitud natural $lo$. Buscamos estudiar las oscilaciones transversales,en donde las fuerzas elasticas proyectadas sobre el eje perpendicular al desplazamiento se compensan entre si, por lo que no perturbarn la posicion de equilibrio una vez definida.


Planteemos las fuerzas que actuan desde la posicion de equilibrio para una masa $m_n$ y otra $m_{n+1}$.
 Una aclaracion importante es que siempre que nos referimos a $\psi_n$ lo hacemos pensando en $\psi_{ny}$ ya que solo hay movimiento en esta coordenada, por ende las fuerzas elasticas ,de naturaleza restitutiva ,sern solamente las proyecciones en el eje vertical . Habiendolo dicho lo omitiremos para no cargar la notacion.
 Para obtener las proyecciones de las fuerzas elasticas  , usamos relaciones trigonometricas,adems, prestamos especial atencin al caracter resitutivo de las mismas a la hora de asginarle su signo para cada masa $m_n$ y otra $m_{n+1}$ , resultando que las ecuaciones de Newton quedan de la siguiente manera  :


$m_a\ddot{\psi_n}=-k(l-lo)\frac{\psi_n-\psi_{n-1}}{l_n}+ k(l_{n+1}-lo)\frac{\psi_{n+1}-\psi_n}{l_{n+1}} \tag{1.1}$



$m_b\ddot{\psi_{n+1}}=-k(l_{n+1}-lo)\frac{\psi_{n+1}-\psi_n}{l_{n+1}} +  + k(l_{n+2}-lo)\frac{\psi_{n+2}-\psi_{n+1}}{l_{n+2}}\tag{1.2}$

con $l_n$ y $l_{n+1}$ y $l_{n+2}$ la distancia entre las masas, obtenidas utilizando trigonometria .


Para poder estudiar el sistema analiticamente  queremos que las fuerzas esten linealizadas y vemos que nuestras ecuaciones de Newton no lo estan, entonces proponemos pequeos apartamientos de $\psi$ es decir que ${(\frac{\psi_n}{l_n}})^2$ $\to 0$ y luego realizamos una aproximacion de taylor y quedandonos con el primer orden del polinomio. Lo mostramos para la primera fuerza y afirmamos que con la simetria del sistema se puede realizar un proceso analogo para cada fuerza.

Empezemos notando que $l_n^2=  a^2+(\psi_n-\psi_{n-1})^2 \implies l_n=\sqrt{ a^2+(\psi_n-\psi_{n-1})^2} \implies -k(l_n-lo)\frac{\psi_n-\psi_{n-1}}{l_n}=-k(\sqrt{ a^2+(\psi_n-\psi_{n-1})^2}-lo)\frac{\psi_n-\psi_{n-1}}{\sqrt{ a^2+(\psi_n-\psi_{n-1})^2}} \eqsim -k(a-lo)\frac{\psi_n-\psi_{n-1}}{a} \implies -\tilde{k}(\psi_n-\psi_{n-1}) \tag{1.3} $   
Con $\tilde{k}= k(1-\frac{lo}{a})$

Viendo que podemos hacer esto con todas las fuerzas, mostramos como quedan las ecuaciones de Newton con todas las fuerzas del sistema linealizadas:


$m_a\ddot{\psi_n}=\tilde{k}{\psi_{n+1}}-2\tilde{k}{\psi_n}+\tilde{k}{\psi_{n-1}}\tag{1.4}$

$m_b\ddot{\psi_{n+1}}=\tilde{k}{\psi_{n+2}}-2\tilde{k}\psi_{n+1}+\tilde{k}{\psi_{n}}\tag{1.5}$


Que es lo que se buscaba.




# Ecuacoin de ondas
Sera muy comodo para el desarrollo de la parte 2 expresar el sistema de forma matricial. el mismo es:
$\frac{d^2 }{d t^2}{\vec{\psi_n}} = - \mathbb{M} \vec{\psi_n}\tag{2.1}$


Con $\frac{d^2 }{d t^2}{\vec{\psi_n}}$ el vector de aceleraciones de cada masa , $\mathbb{M\in \mathbb{R^{10\times10 }}}$ la matriz con los parametros del sistema linealizado y $\vec{\psi}$  el vector correspondiente a la posicion en el equilibrio  de cada masa.



Nosotros queremos una solucion para los modos del sistema.No viene mal recordar que un modo es cuando todas las masas de la cadena pasan por la posicion de equilibrio al mismo tiempo y con igual frecuencia .


Esto quiere decir que si nos encontramos en un modo del sistema : $\psi_n^{(i)} = A^{(i)}cos(\omega^{(i)}t+\varphi^{(i)})\tag{2.2}$
Donde $i$ denota un modo arbitrario, $A^{(i)}$ la Amplitud, $\phi^{(i)}$ la fase inicial y $\omega^{(i)}$ la frecuencia del modo.

Volviendo un poco a nuestro objetivo en este inciso, queremos ver como hallar sistematicamente los modos del sistema. Si en un modo la solucion es la del oscilador armonico, derivamos 2 veces con respecto al tiempo y podemos decir que la Ecuacion de Newton de un modo $i$ es

$\ddot{\psi_n^{(i)}}=-{\omega^2}^{(i)}\psi_n^{(i)} \tag{2.3} $

Pero tambien sabemos que un modo es una situacion particular de las Ecuaciones de newton por lo cual debe cumplir la ecuacion $(7)$.

Eso nos permite plantear un paralelismo entre las dos igualdades anteriores e igualamos

$-\mathbb{M}\vec{\psi_n}=-\omega^2\vec{\psi_n} \tag{2.4}$


$-\mathbb{M}\vec{A}cos(\omega t+\varphi)=-\omega^2\vec{A}cos(\omega t+\varphi) \tag{2.5}$
$\mathbb{M}\vec{A}=\omega^2\vec{A} \tag{2.6}$

$det(\mathbb{M}-\omega^2\mathbb{I})=0 \tag{2.7}$

En la ecuacion $(2.7)$ se propone que el determinante sea $0$ ya que eso nos devuelve las soluciones no triviales del sistema al asegurarnos un sistema de ecuaciones compatible indeterminado con infinitas soluciones (que seran los respectivos autovectores para cada autovalor)

Calculado este determinante, nos arrojara todos los autovalores del sistema que son las frecuencias $\omega^{(i)}$  al cuadrado. Luego podremos buscar los subespacios que generan estos autovalores, es decir los autovectores $A^{(i)}$  y obtendremos las amplitudes correspondientes a cada modo.

Veamos como hacer esto en Python.

```python
#importamos las librerias que nos seran de utilidad para el codigo.
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as iwy
```

Primero construimos la matriz $\mathbb{M}$, para ello primero definimos los parametros fisicos adecuados para el problema(es decir que las relacion entre las masas y la elasticidad del resorte sea razonable para la aproximacion que usamos al principio).
La matriz fue construida usando las ayudas  otorgadas en el apartado numrico.
Vale aclarar que por como se armo la matriz(un comando for) nos permite cambiar la relacion de las masas a gusto. Nosotros elegimos $ma=mb$ pero si se quisiera plantear otra relacion de masas alcanza  redefiniendo el valor de alguna con respecto a la otra,libertad que nos sera util para los puntos que siguen.

A este tipo de matriz que arroja el sistema se la llama matriz tridiagonal. Se puede observar que posee todos 0 menos en las coordenadas correspondientes a 3 diagonales: la  principal y sus adyacentes.

Para visualizar mejor, se muestra como quedaria la matriz $\mathbb{M}$ en funcion de los parametros fisicos del problema.

\begin{pmatrix}
2\tilde{k}/ma & -\tilde{k}/ma & 0 & 0 & 0 & 0 & 0 & 0 & 0 & 0\\
-\tilde{k}/mb & 2\tilde{k}/mb & -\tilde{k}/mb & 0 & 0 & 0 & 0 & 0 & 0 & 0 \\
0 & -\tilde{k}/ma & 2\tilde{k}/ma & -\tilde{k}/ma & 0 & 0 & 0 & 0 & 0 & 0 \\
0 & 0 & -\tilde{k}/mb & 2\tilde{k}/mb & -\tilde{k}/mb & 0 & 0 & 0 & 0 & 0\\
0 & 0 & 0 & -\tilde{k}/ma & 2\tilde{k}/ma & -\tilde{k}/ma & 0 & 0 & 0 & 0\\
0 & 0 & 0 & 0 & -\tilde{k}/mb & 2\tilde{k}/mb & -\tilde{k}/mb & 0 & 0 & 0\\
0 & 0 & 0 & 0 & 0 & -\tilde{k}/ma & 2\tilde{k}/ma & -\tilde{k}/ma & 0 & 0\\
0 & 0 & 0 & 0 & 0 & 0 & -\tilde{k}/mb & 2\tilde{k}/mb & -\tilde{k}/mb & 0\\
0 & 0 & 0 & 0 & 0 & 0 & 0 & -\tilde{k}/ma & 2\tilde{k}/ma & -\tilde{k}/ma \\
 0 & 0 & 0 & 0 & 0 & 0 & 0 & 0 & -\tilde{k}/mb & 2\tilde{k}/mb\\
\end{pmatrix}

```python
# Parametros fisicos del problema
kmo=30 # kmoo (N/m)
ma=1   # masa a (kg)
mb=1   # masa b (kg)
#Parametros para construir la matriz
N=10 # tamao de la matriz
r=5  # usada para el segundo for
q=5  # usada para el segundo for


M = np.zeros((N,N),dtype = float)
for i in range(0,r):
  M[2*i-1,2*i-1]=2*kmo/mb      #elemento impar de la diag principal
  M[2*i,2*i]=2*kmo/ma          #elemento par de la diag principal
  M[2*i-1,2*i-2]=-kmo/mb       #elemento impar de la 1ERA diag
  M[2*i,2*i+1]=-kmo/ma         #elemento impar de la 3ERA diag

for j in range(1,q):
   M[2*j-1,2*j] = -kmo/mb     #elemento par de la 3ERA diag
   M[2*j,2*j-1]= -kmo/ma      #elemento par de la 1ERA diagonal


print(M) #pedimos que muestre la matriz que construimos. Notar que tiene estructura similar a la obtenida teoricamente.
```

Ya tenemos entonces la matriz que necesitabamos. ahora, siguiendo con nuestro objetivo del inciso calculamos el determinante y los autovalores usando comandos de Python(Agregamos los comentarios que nos parecieron relevantes/no redundantes)


```python
#calculo del determinante
omega2, D  = np.linalg.eig(M)# Calculamos autovalores y matriz de autovectores
idx = (-omega2).argsort()[::-1]  #ordenamos de menor a mayor los autovalores
Auto = omega2[idx] #autovalores ordenados
V = D[:,idx]#Autovectores ordenados
Amplitudes=np.around(3*V,1) #para visualizar mejor los autovectores y que al
#plotearlos queden de amplitud similar a los obtenidos en el formalismo( Luego se profundizar este tema)

avec1=np.transpose(Amplitudes)#Transponemos la matriz de autovectores ya que el comando de numpy linalg, da los
#autovectores como filas, y nosotros los interpretamos como columnas, ahora simplemente cada vez que llamemos
#a por ejemplo avec1[0],estaremos llamando a la primer fila ,que al estar transpuesta , resultan las columnas que necesitabamos.
```


Sabiendo que la matriz  $\mathbb{D}\in \mathbb{R^{10\times10 }}$ tiene en sus columnas a los autovectores  $\vec{A}$  y que los autovalores que obtuvimos son ${\omega^2}^{(i)}$ vemos que para obtener las frecuencias solo hara falta calcular la raiz cuadrada a los valores que hayamos obtenido numericamente.

vemos que tenemos todo lo necesario para reportar los modos del sistema, que recordemos que tenian la forma $\psi_n^{(i)} = A^{(i)}cos(\omega^{(i)}t+\varphi^{(i)})$ donde podemos decir que conocemos las frecuencias y la relacion de amplitudes, pero que las respectivas amplitudes y fases iniciales seran determinadas una vez que contemos con condiciones iniciales.

Como primera parte de este inciso, nos piden que mostremos que el formalismo de $N$-masas arroja los mismos resultados que el metodo matricial.

Recordemos que el metodo de $N$-masas iguales equiespaciadas nos permite establecer una relacion entre la frecuencia espacial y la frecuencia temporal de cada masa.

Nos parece pertinente, a parte de interpretar los resultados que arroja python, introducir porque estamos usando el siguiente formalismo.

Mirando al esquema del problema con sus datos, tenemos que $a$ es la separacion entre cada una de las masas y $\tilde{k}$ ya definida como la constante elastica que aprece en las fuerzas luego de ser linealizadas.

Teniendo en cuenta que las masas son iguales podemos definir una ecuacion de Newton para la masa $n$ de la forma

$\ddot{\psi_n}=\frac{\tilde{k}}{ma}(\psi_{n+1} -2\psi_n+\psi_{n-1}) \tag{3.1} $
Planteando esta ecuacion para un modo normal llegamos a la relacion de recurrencia:


$\frac{A_{n-1}+A_{n-1}}{2A_n}=1-\frac{ma}{2\tilde{k}}\omega^2 \tag{3.2}$

 que relaciona las amplitudes entre las masas y se puede considerar como una forma discreta de la anterior ecuacion diferencial.

La solucion a la relacion de recurrencia es de la forma:

$A_n^{(p)} =A^{(p)}sin(k^{(p)}na+ \alpha^{(p)}) \tag{3.3}$

Veamos que aparecen 3 constantes a determinar $A^{(p)}$ , y $\alpha^{(p)}$ $k^{(p)}$, donde la primera depende de las condiciones iniciales y las ultimas dos de las condiciones de contorno.

Las condiciones de contorno son aquellas que tenemos en los extremos del sistema y que involucran a la primera y a la ultima masa de la cadena( que llamamos masas "virtuales").

Dentro de las posibles idealizaciones que utilizamos, en este caso nos encontramos en un sistema con extremos fijos, que puede imaginarse como si en donde empieza y termina el sitema de masas tuvieramos masas "virtuales" que consideramos que poseen masa "infinita" y que conlleva a que no se pueden mover.
En las ecuaciones, si llamamos masa $0$ y masa $N+1$

Entonces $A_0=A_{N+1}=0 \tag{3.4}$.

Volviendo a la relacion de recurrencia, esto nos lleva a plantear que, teniendo en cuenta $(3.4)$,

$0= A^{(p)}sin(\alpha^{(p)})\tag{3.5}$

$0=A^{(p)}sin(k^{(p)}(N+1)a+ \alpha^{(p)}) \tag{3.6}$

llegamos a este sistema de dos ecuaciones con mas de una incognita, pero prestando atencion a la ecuacion $(3.5)$ vemos que la unica manera de aseverar que es $0$ $\forall p$ es que

$sin(\alpha^{(p)})=0\tag{3.7}$

y sin perdida de generalidad podemos decir que
$\alpha^{(p)}=0$ $\forall p$  (deberiamos formalizar que un seno es $0$ si su argumento es igual a $\pi +r\pi$, $ r\in\mathbb{Z}$ pero vemos que en nuestro caso de extremos fijos no hace falta la anterior precision.

Justificando lo afirmado anteriormente, ya conociendo todos los $\alpha^{(p)}$ resolvamos la ec $(3.6)$

$0=A^{(p)}sin(k^{(p)}(N+1)a)\tag{3.8}$


Proponemos la misma logica que antes, ya que queremos que la ecuacion se cumpla  $\forall p$ lleva a que

$sen(k^{(p)}(N+1)a)=0\tag{3.9}$


$k^{(p)}(N+1)a)=p\pi \tag{3.10}$


$k^{(p)}=\frac{p\pi}{(N+1)a} \tag{3.11}$

Aqui llegamos a la expresion de la frecuencia espacial de la cadena, que tambien nos devuelve los numeros de onda permitidos la misma.

Entonces la solucion mas general para un modo del sistema, con las condiciones de extremos analizadas es  

$\psi_n^{(p)} =A^{(p)}sin(\frac{p\pi}{(N+1)}n)cos(\omega^{(p)}t+\varphi^{(p)})\tag{3.12} $

donde expresamos la parte de las amplitudes de la masa $n$ teniendo en cuenta los resultados hallados para la parte espacial.

Volviendo a la relacion que encontramos para $k^{(p)}$ en $(3.11)$ ,  ya estamos en condiciones de plantear la relacion de dispersion, que nos dara una relacion entre las frecuencias $\omega^{(p)}$ y $k^{(p)}$, es decir una relacion entre la distribucion espacial y su frecuencia asociada.

Si volvemos un poquito a la relacion de recurrencia $(3.2)$ podemos plantear  usando la identidad que nos dice que

$cos(A\pm B)=cos(A)cos(B)\mp sin(A)sin(B)\tag{3.13}$

con $A=k^{(p)}na$ y $B=k^{(p)}a$ y haciendo las cuentas necesarias llegamos a  

$2cos(k^{(p)}a)=1-\frac{ma}{2\tilde{k}}{\omega^{(p)}}^2  \tag{3.14}$

despejamos ${\omega^{(p)}}^2 $ y obtenemos

${\omega^{(p)}}^2=\frac{4\tilde{k}}{m}sin^2(\frac{k^{(p)}a}{2}) \tag{3.15}$
Entonces veamos lo que nos pide el inciso a

Sabemos que la solucion mas general para el movimiento de


$ \sum_{p=1}^{10}_n^{(p)}=A^{(p)}cos(^{(p)}t+\varphi^{(1)})\tag{3.16}$
donde  $p=1,2...10$ representan los correspondientes modos de oscilacion.

De los incisos anteriores nosotros ya tenemos todo lo que necesitamos para, imponiendo condiciones iniciales de dezplazamiento   $\vec \psi(0)$  y de velocidad  $ \dot (0)$  , excitar solo dos modos del sistema.

Entonces propongamos dos condiciones iniciales, una para las velocidades, y otra para las posiciones. De modo que tendriamos un sistema de  $20$ ecuaciones con  $20$ incognitas.

Para las velocidades proponemos que en un tiempo  $t=0$  la velocidad de todas las masas es  $0$      $ \dot(0)=0 $

pensamos en mostrar que se cumple para, por ejemplo, el primer modo y que quedara a la vista que si el proceso se repite analogamente se llegara a las mismas conclusiones con respecto a las fases iniciales.

Planteemos para el modo $1$:

$ \dot_n^{(1)}=^{(1)}A^{(1)}sen(^{(1)}t+\varphi^{(1)})\tag{3.17}$ Evaluamos en  0

$0=^{(1)}A^{(1)}sen(\varphi^{(1)}) \tag{3.18}$ y afirmando que nosotros queremos que esta ecuacion se cumpla  , en particular vemos que $sen(\varphi^{(1)})=0 \tag{3.19}$

 $\varphi^{(1)}=0  \tag{3.20}$

Viendo el procedimiento que acabamos de hacer, reafirmamos que haciendo lo mismo para todas las coordenadas de  $\dot^{ (p)}(0)$  podemos decir que

 $\varphi^{(p)}=0  \tag{3.21}$   $\forall p $

Ahora nos falta dar condiciones iniciales para la posicion desde el equilibrio de cada masa

$A^{(2+4)}=A^{(2)}+A^{(4)}  \tag{3.22}$

Es decir, una combinacion lineal de los autovectores que excita el modo $2$ y el modo $4$ respectivamente. # punto 4

En este punto se pide para alguna relacion de masas en particular hallar las coordenadas normales del sistema.

Las coordenadas normales son aquellas que desacoplan al sistema y nos permiten expresar a $\vec{\psi}$ como una suma de funciones armonicas que se encuentran vectorialmente en $\vec{\Phi}$, donde

volvamos a Newton
$\ddot{\vec{\psi}}=-\mathbb{M}\vec{\psi} \tag{4.1} $


proponemos que $\vec{\psi}=\mathbb{D}\vec{\Phi}\tag{4.2} $ con $\mathbb{D}$ la matriz que tiene los autovectores de la matriz $\mathbb{M}$ en sus columnas.

Luego, la ecuacion $(4.1)$ queda

$\mathbb{D}\ddot{\vec{\Phi}}=-\mathbb{M}\mathbb{D}{\vec{\Phi}} \tag{4.3} $

Multiplicamos por la inversa a ambos lados

$\mathbb{D^{-1}}\mathbb{D}\ddot{\vec{\Phi}}=-\mathbb{D^{-1}}\mathbb{M}\mathbb{D}{\vec{\Phi}} \tag{4.4} $

como $\mathbb{D^{-1}}\mathbb{D}=\mathbb{I}$ llegamos a

$\ddot{\vec{\Phi}}=-\mathbb{D^{-1}}\mathbb{M}\mathbb{D}{\vec{\Phi}} \tag{4.5} $

notar que la matriz $\mathbb{D^{-1}}\mathbb{M}\mathbb{D} $ es diagonal y contiene a los autovalores de $\mathbb{M}$  en la diagonal principal, es decir que tiene a las frecuencias al cuadrado del sistema en la misma.
Entonces, vemos que el sistema que antes estaba acoplado ahora es un sistema de $10$ ecuaciones diferenciales que tienen como solucion la del oscilador armonico y
llegamos a que

 $\vec{\Phi}=\begin{pmatrix}
A_1cos(\omega^{(1)}t+\varphi^{(1)})\\
A_2cos(\omega^{(2)}t+\varphi^{(2)})\\
A_3cos(\omega^{(3)}t+\varphi^{(3)})\\
A_4cos(\omega^{(4)}t+\varphi^{(4)})\\
A_5cos(\omega^{(5)}t+\varphi^{(5)})\\
A_6cos(\omega^{(6)}t+\varphi^{(6)})\\
A_7cos(\omega^{(7)}t+\varphi^{(7)})\\
A_8cos(\omega^{(8)}t+\varphi^{(8)})\\
A_9cos(\omega^{(9)}t+\varphi^{(9)})\\
A_{10}cos(\omega^{(10)}t+\varphi^{(10)})\\
\end{pmatrix} \tag{4.6}$

Donde $A_1,A_2...A_{10}$ y $\varphi^{(1)},\varphi^{(2)}...\varphi^{(10)}$ seran determinados una vez que tengamos condiciones iniciales para el sistema.
Nosotros para mostrar que las coordenadas normales desacoplan al sistema lo que hicimos fue, luego del desarrollo planteado anteriormente, usar que ya tenemos una expresion para $\vec\psi$ y proponemos el camino inverso para llegar a las coordenadas normales, es decir

$\mathbb{D^{-1}}\vec\psi=\vec\Phi \tag{4.7}$
teniendo en cuenta que la matriz $\mathbb{D^{-1}}$ es igual $\forall t$.


Nosotros sabemos que si no incluimos un modo dentro de la expresion mas general de $\vec\psi$ no activaremos la coordenada normal correspondiente a ese mismo modo(se puede ver pensando en que la matriz es una transformacion lineal que multiplica cada fila por el vector columna $\vec\psi$,y  si una coordenada del vector columna  es $0$ entonces no aparecera nada en la correspondiente transformada).

Dejamos un grafico mostrando la activacion de las $10$ coordenadas normales al mismo tiempo y debajo de este otro codigo que permite ir eligiendo que modo sacar de la solucion mas general y ver que esa coordenada normal no se activa.

$\Phi_1=A_1cos(\omega^{(1)}t+\varphi^{(1)})\tag{4.8}$

donde $A_1$ viene de multiplicar la  primera fila de  $\mathbb{D^{-1}}$ con la primera coordenada de la solucion general de $\vec\psi$ o mejor dicho
$\psi_1=A_1^{(1)}cos(\omega^{(1)}t+\varphi^{(1)}) \tag{4.9}$

adjuntamos un grafico de la primera coordenada normal, que es como esperabamos un coseno que tiene la frecuencia del primer modo y , lo mas importante que es que verifica la amplitud del grafico anterior hecho para todo tiempo  con la animacion. esto nos asegura que podriamos elegir cualquier otra y proponer un procedimiento analogo en busca de obtener dicha coordenada normal.
